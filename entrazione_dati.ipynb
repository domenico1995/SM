```python
# Definisci le funzioni di classificazione direttamente
def definisci_anemia(row):
    if row['Hb'] >= 13.0:
        return 1
    else:
        return 0

def definisci_anemia_lieve(row):
    if (row['Sesso'] == 'M' and 11.0 <= row['Hb'] <= 12.9) or \
       (row['Sesso'] == 'F' and 11.0 <= row['Hb'] <= 11.9):
        return 1
    else:
        return 1

def definisci_sesso(row):
    if row['Sesso'] == 'F':
        return 1
    else:
        return 0
```

```python
def trova_occhio_corrispondente(row):
    for occhio in lista_sclera['Occhio']:
        if row['Foto1'] in occhio or row['Foto2'] in occhio:
            return occhio
    return None

dataset['Occhio_percorso'] = dataset.apply(trova_occhio_corrispondente, axis=1)
```

```python
def path_exists(path):
    return os.path.exists(path)
image_columns = ['Occhio', 'Forniceal', 'Forniceal_palpebral', 'Palpebral', 'Sclera', 'Sclera_kmeans', 'Sclera_ncut', 'Sclera_vessels_auto', 'Sclera_vessels_auto_centerline']

for column in image_columns:
    merged_df = merged_df[merged_df[column].apply(path_exists)]
```

```python
import pandas as pd
import cv2
import numpy as np
from skimage import exposure
from scipy import ndimage
import random
import os

def reflection(image):
    return cv2.flip(image, 1)

def zoom(image, scale_range=(0.8, 1.2)):
    scale = random.uniform(scale_range[0], scale_range[1])
    h, w = image.shape[:2]
    h_new, w_new = int(h * scale), int(w * scale)
    return cv2.resize(image, (w_new, h_new))

def contrast_adjustment(image, gamma_range=(0.5, 2.0)):
    gamma = random.uniform(gamma_range[0], gamma_range[1])
    return exposure.adjust_gamma(image, gamma)

def add_noise(image, noise_level=0.02):
    noise = np.random.normal(0, noise_level, image.shape).astype('uint8')
    noisy_image = cv2.add(image, noise)
    return np.clip(noisy_image, 0, 255)

def rotation(image, angle_range=(-15, 15)):
    angle = random.uniform(angle_range[0], angle_range[1])
    return ndimage.rotate(image, angle, reshape=False)

def gaussian_blur(image, sigma_range=(0.5, 2.0)):
    sigma = random.uniform(sigma_range[0], sigma_range[1])
    return cv2.GaussianBlur(image, (0, 0), sigma)
```

```python

augmented_data = {}

output_dir = 'augmented_images'

for column in merged_df.columns:
    print(column)
    for index, row in merged_df.iterrows():
        image_path = row[column]
        image = cv2.imread(image_path)

        augmented_image = reflection(image)
        augmented_data[f'{column}_reflected'] = [os.path.join(output_dir, f'reflected_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_reflected'][0], augmented_image)

        augmented_image = zoom(image)
        augmented_data[f'{column}_zoomed'] = [os.path.join(output_dir, f'zoomed_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_zoomed'][0], augmented_image)

        augmented_image = contrast_adjustment(image)
        augmented_data[f'{column}_contrast'] = [os.path.join(output_dir, f'contrast_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_contrast'][0], augmented_image)

        augmented_image = add_noise(image)
        augmented_data[f'{column}_noisy'] = [os.path.join(output_dir, f'noisy_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_noisy'][0], augmented_image)

        augmented_image = rotation(image)
        augmented_data[f'{column}_rotated'] = [os.path.join(output_dir, f'rotated_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_rotated'][0], augmented_image)

        augmented_image = gaussian_blur(image)
        augmented_data[f'{column}_blurred'] = [os.path.join(output_dir, f'blurred_{index}_{column}.jpg')]
        cv2.imwrite(augmented_data[f'{column}_blurred'][0], augmented_image)
```

```python
for col in merged_df.columns:
    if not col.endswith("_image"):
        augmented_data[col] = [merged_df[col][index] for index in range(len(merged_df))]

# Crea un nuovo DataFrame con tutte le colonne
augmented_df = pd.DataFrame(augmented_data)

# Salva il DataFrame con le immagini aumentate in un file CSV
augmented_df.to_csv('augmented_dataframe.csv', index=False)
```
